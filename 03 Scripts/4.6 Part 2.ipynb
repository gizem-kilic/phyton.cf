Looking at the overview of both dataframes, I can see that product_id is a mutual column. Let's try inner join.

```python
df_merged.merge(df_prds, on='product_id', how='inner', indicator='merge_indicator')
```

Confirm the results of the merge using the merge flag.


```python
merged_df = df_merged.merge(df_prds, on='product_id', how='inner', indicator='merge_indicator')
```

```python
merged_df['merge_indicator'].value_counts()
```

```python
merged_df_outer = df_merged.merge(df_prds, on='product_id', how='outer', indicator='merge_indicator')
```

```python
merged_df_outer['merge_indicator'].value_counts()
```

As we can see the different values under merge_indicator with the outer join, we can say that there is no full match between the two dataframes.

```python
# export the data in pickle format
merged_df.to_pickle(os.path.join(r"C:\Users\gkgiz\Desktop\CF\Achievement 4\09 - 2024 Instacart Basket Analysis\02 Data\Prepared Data", 'ords_prods_merge.pkl'))
```
